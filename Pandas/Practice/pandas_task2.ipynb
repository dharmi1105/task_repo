```python
import pandas as pd
import json
```

```python
data1 = pd.DataFrame({
    'name' : ['Amy','Robert','Rahul'],
    'Result':[50,30,40],
    'id' : [1,2,3]
},
index = [1,2,3]
)

data2 = pd.DataFrame({
    'name' : ['Jolly','Paul','Rachana'],
    'Result':[10,30,60],
    'id' : [1,2,3]
},
index = [4,5,6]
)

result = pd.merge(data1,data2, on = "id")
result
```

```python
data3 = pd.DataFrame({
    'continent' : ['Asia','Africa','America','Europe','Antarctica'],
    'countries': ['India','South Africa','USA','Ireland','Antarctica'],
    'pop': [102,105,107,106,33]
})

data4 = pd.DataFrame({
    'continent' : ['Asia','Africa','America','Europe','Antarctica'],
    'countries': ['Malasiya','South Africa','Canada','London','Antarctica'],
    'pop': [100,105,80,130,33]
})
```

```python
result = pd.merge(data3,data4, how = 'left', on = 'continent')
result
```

```python
result = pd.merge(data3,data4, how = 'right', on = 'continent')
result
```

```python
result = pd.merge(data3,data4, how = 'inner', on = 'continent')
result
```

```python
result = pd.merge(data3,data4, how = 'outer', on = 'continent')
result
```

It will be chaos if we use cross join and the results are not correct in some cases just executed for example case

```python
result = pd.merge(data3,data4, how = 'cross')
result
```
